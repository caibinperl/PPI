```python
import os
import numpy as np
import torch
import h5py
from embed import seq2tensor
```

```python
spe = "yeast"

seed = 1234

seq_file = os.path.join(spe, "seq/protein.dictionary.tsv")
emb_file = "embedding/vec5_CTC.txt"
action_file = os.path.join(spe, "actions/protein.actions.tsv")
embedding_h5 = os.path.join(spe, "seq/pipr.embedding.h5")

torch.manual_seed(seed)
np.random.seed(seed)
```

```python
seq2t = seq2tensor(emb_file)
id2seq = {}

with open(seq_file, "r") as fin, h5py.File(embedding_h5, "w") as h5fout:
    for line in fin:
        line = line.strip()
        if not line:
            continue
        ss = line.split("\t")
        id = ss[0]
        seq = ss[1]
        id2seq[id] = seq
        x = seq2t.embed(seq)
        if id not in h5fout:
            dset = h5fout.require_dataset(
                id,
                shape=x.shape,
                dtype="float32",
                compression="lzf",
            )
            dset[:] = x
```
